```python
# File statistics

def file_stats(directory):
    file_sizes = []
    extensions = set()
    
    for subdir, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(subdir, file)
            file_size = os.path.getsize(file_path)
            file_sizes.append(file_size)
            extensions.add(os.path.splitext(file)[1].lower())
    
    total_files = len(file_sizes)
    total_avg_size = sum(file_sizes) / total_files
    max_size = max(file_sizes)
    min_size = min(file_sizes)
    
    print("All file types found :", extensions)
    print("Total number of files:", total_files)
    print("Average file size    : {:.2f} KB".format(total_avg_size / 1024))
    print("Maximum file size    : {:.2f} KB".format(max_size / 1024))
    print("Minimum file size    : {:.2f} KB".format(min_size / 1024))
    
    return file_sizes

# Get file statistics for the training directory
file_sizes = file_stats(train_dir)
```

```python
def calculate_brightness(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    brightness = np.mean(gray)
    return brightness

def calculate_contrast(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    contrast = np.std(gray)
    return contrast

def analyze_brightness_contrast(root_dir):
    results = []
    for class_name in os.listdir(root_dir):
        class_dir = os.path.join(root_dir, class_name)
        if os.path.isdir(class_dir):
            for file_name in os.listdir(class_dir):
                file_path = os.path.join(class_dir, file_name)
                if os.path.isfile(file_path):
                    img = cv2.imread(file_path)
                    if img is not None:
                        brightness = calculate_brightness(img)
                        contrast = calculate_contrast(img)
                        results.append({'Class': class_name, 'File': file_path, 'Brightness': brightness, 'Contrast': contrast})
                    else:
                        print(f"Error: Unable to read image '{file_path}'")
    return results

def print_statistics_and_display_images(root_dir, dataset_name):
    # Analyze brightness and contrast
    results = analyze_brightness_contrast(root_dir)
    
    # Print statistics
    df = pd.DataFrame(results)
    df_stats = df.groupby('Class').agg({'Brightness': ['mean', 'std', 'min', 'max'], 
                                        'Contrast': ['mean', 'std', 'min', 'max']}).reset_index()
    df_stats.columns = ['Class', 'Brightness Mean', 'Brightness Std', 'Brightness Min', 'Brightness Max', 
                        'Contrast Mean', 'Contrast Std', 'Contrast Min', 'Contrast Max']
    display(df_stats)
    
    # Sort results by brightness and contrast
    results.sort(key=lambda x: x['Brightness'])
    lowest_brightness_images = [cv2.cvtColor(cv2.imread(img['File']), cv2.COLOR_BGR2RGB) for img in results[:10]]
    highest_brightness_images = [cv2.cvtColor(cv2.imread(img['File']), cv2.COLOR_BGR2RGB) for img in results[-10:]]
    
    results.sort(key=lambda x: x['Contrast'])
    lowest_contrast_images = [cv2.cvtColor(cv2.imread(img['File']), cv2.COLOR_BGR2RGB) for img in results[:10]]
    highest_contrast_images = [cv2.cvtColor(cv2.imread(img['File']), cv2.COLOR_BGR2RGB) for img in results[-10:]]

# Print statistics and display images for the train dataset
print_statistics_and_display_images(train_dir, "Train")
```

```python
# Identify the RGB channel distribution of the images from each class
def calculate_rgb_distribution(root_dir):
    rgb_distribution_per_class = {}
    for class_name in os.listdir(root_dir):
        class_dir = os.path.join(root_dir, class_name)
        if os.path.isdir(class_dir):
            class_rgb_distribution = []
            for file_name in os.listdir(class_dir):
                file_path = os.path.join(class_dir, file_name)
                if os.path.isfile(file_path):
                    img = cv2.imread(file_path)
                    if img is not None:
                        # Ensure image has all three channels (RGB)
                        if img.shape[2] == 3:
                            # Calculate RGB distribution
                            b, g, r = cv2.split(img)
                            b_mean = b.mean()
                            g_mean = g.mean()
                            r_mean = r.mean()
                            class_rgb_distribution.append((b_mean, g_mean, r_mean))
                        else:
                            print(f"Skipping file '{file_path}' due to missing RGB channels")
                    else:
                        print(f"Error: Unable to read image '{file_path}'")
            if class_rgb_distribution:
                # Calculate average RGB values for the class
                avg_b = sum([b[0] for b in class_rgb_distribution]) / len(class_rgb_distribution)
                avg_g = sum([g[1] for g in class_rgb_distribution]) / len(class_rgb_distribution)
                avg_r = sum([r[2] for r in class_rgb_distribution]) / len(class_rgb_distribution)
                rgb_distribution_per_class[class_name] = {'R Mean': avg_r, 'G Mean': avg_g, 'B Mean': avg_b}
    return rgb_distribution_per_class

# Assuming your Jupyter Notebook is in the same directory as the train folder
current_dir = '.'
train_dir = os.path.join(current_dir, 'New Plant Diseases Dataset(Augmented)/train')

# Calculate RGB distribution for train dataset per class
train_rgb_distribution_per_class = calculate_rgb_distribution(train_dir)

# Create DataFrames for train dataset
train_df = pd.DataFrame.from_dict(train_rgb_distribution_per_class, orient = 'index')

# Display RGB distribution table for train dataset
display(train_df)
```

# **4.0 Deep Learning Model Implementation**

## **4.1 ResNet50-BiLSTM (4 Layers)**

```python
# Load ResNet50 as feature extractor
feature_extraction_ResNet50 = ResNet50(weights = 'imagenet', 
                                       include_top = False, 
                                       input_shape = (224, 224, 3))
feature_extraction_ResNet50.trainable = False # Freeze the layers 

# Define the ResNet50-BiLSTM model architecture
ResNet50_BiLSTM_model = Sequential([
    # Input (Feature Extraction) Layer - ResNet50 extract features (height, width, number of feature channels) from the image 
    feature_extraction_ResNet50,
    
    # Hidden Layers
    # Time Distributed Layer - Convert the extracted features of ResNet50 (time steps, height, width, number of feature channels)
    #                          into a 2D output (time steps, flat vector features) which can be processed by the BiLSTM layer
    TimeDistributed(Flatten(),
                    name = 'Time_Distributed_Layer_1'), 
    # BiLSTM layer
    Bidirectional(LSTM(units = 128, # Dimensionality of the output space with bidirectional operation
                       return_sequences = False),
                       name = 'BiLSTM_Layer_1'),
    
    # Output layer for multiple classification
    Dense(units = 10, # Dimensionality of the output space
          activation='softmax',
          name = 'Dense_Layer_1')
], name = 'ResNet50_BiLSTM_4_Layers')

# ResNet50-BiLSTM model summary
ResNet50_BiLSTM_model.summary(line_length = 100)
```

# **5.0 Deep Learning Model Hyperparameter Tuning**

## **5.1 ResNet50-BiLSTM (9 Layers) Model Hyperparameter Tuning (Hyperband)**

```python
# Function to define the list of ResNet50-BiLSTM hyperparameters for hyperparameters tuning
def ResNet50_BiLSTM_model_tuning_function(hp):
    # Load ResNet50 as feature extractor
    feature_extraction_ResNet50 = ResNet50(weights = 'imagenet', 
                                           include_top = False, 
                                           input_shape = (224, 224, 3))
    feature_extraction_ResNet50.trainable = False # Freeze the layers 

    # Define the ResNet50-BiLSTM model architecture with the list of hyperparameters
    ResNet50_BiLSTM_model_tuning = Sequential([
        # Input (Feature Extraction) Layer - ResNet50 extract features (height, width, number of feature channels) from the image 
        feature_extraction_ResNet50,
        
        # Hidden Layers
        # Time Distributed Layer - Convert the extracted features of ResNet50 (time steps, height, width, number of feature channels)
        #                          into a 2D output (time steps, flat vector features) which can be processed by the LSTM layer
        TimeDistributed(Flatten(),
                        name = 'Time_Distributed_Layer_1'), 
        # Batch Normalization Layer - Normalize the flat vector features that are input to the BiLSTM
        BatchNormalization(name = 'Batch_Normalization_Layer_1'),
        # BiLSTM layer
        Bidirectional(LSTM(units = hp.Choice('BiLSTM_units_1', values = [16, 32, 64, 128, 256]), # Dimensionality of the output space with bidirectional operation
                           return_sequences = True),
                           name = 'BiLSTM_Layer_1'),
        # Dropout layer - Disregarding certain nodes in a layer at random during training
        Dropout(rate = 0.65, # Fraction of the inputs to drop
                name = 'Dropout_Layer_1'), 
        # Batch Normalization Layer - Normalize the flat vector features that are input to the BiLSTM
        BatchNormalization(name = 'Batch_Normalization_Layer_2'),
        # BiLSTM layer
        Bidirectional(LSTM(units = hp.Choice('BiLSTM_units_2', values = [16, 32, 64, 128, 256]), # Dimensionality of the output space with bidirectional operation
                           return_sequences = False),
                           name = 'BiLSTM_Layer_2'),
        # Dropout layer - Disregarding certain nodes in a layer at random during training
        Dropout(rate = 0.65, # Fraction of the inputs to drop
                name = 'Dropout_Layer_2'), 

        # Output layer for multiple classification
        Dense(units = 10, # Dimensionality of the output space
            activation='softmax',
            name = 'Dense_Layer_1')
    ], name = 'Tuned_ResNet50_BiLSTM_9_Layers')

    # Compile the ResNet50-BiLSTM model
    ResNet50_BiLSTM_model_tuning.compile(loss = 'categorical_crossentropy',
                                         optimizer = Adam(learning_rate = 1e-6),
                                         metrics = ['accuracy'])
    
    return ResNet50_BiLSTM_model_tuning
```

```python
# Initialize the ResNet50-BiLSTM tuner
ResNet50_BiLSTM_tuner = kt.Hyperband(
    ResNet50_BiLSTM_model_tuning_function,
    objective = 'val_accuracy',
    max_epochs = 300,
    directory = 'keras_ResNet50_BiLSTM_tuner_hyperband',
    project_name = 'ResNet50_BiLSTM_tuning'
)

# Summary of the ResNet50-BiLSTM hyperparameter search space
ResNet50_BiLSTM_tuner.search_space_summary()
```

```python
# Execute the ResNet50-BiLSTM model hyperparameter hyperband search
ResNet50_BiLSTM_tuner.search(train_generator_resnet50_BiLSTM,
                             epochs = 300,
                             validation_data = val_generator_resnet50_BiLSTM)
```

```python
# Define the custom-based learning rate scheduler class
class Custom_Learning_Rate_Scheduler(Callback):
    def __init__(self, patience = 10, reduce_factor = 10, reduce_every = 5):
        super(Custom_Learning_Rate_Scheduler, self).__init__()
        self.patience = patience # Number of epochs to wait before reducing the learning rate
        self.reduce_factor = reduce_factor # Factor of the learning rate should be reduced
        self.reduce_every = reduce_every # Number of epochs to check if the learning rate needs to be reduced
        self.wait = 0 # Counter to keep track how many epochs have passed after triggered
        self.stored_epoch = -1 # Keeps track of the last epoch

    def on_epoch_end(self, epoch, logs = None):
        logs = logs or {}
        train_acc = logs.get('accuracy')
        val_acc = logs.get('val_accuracy')
        if train_acc > val_acc:
            if self.wait == 0:
                self.stored_epoch = epoch
            self.wait += 1
            if self.wait >= self.patience:
                if (epoch - self.stored_epoch) % self.reduce_every == 0:
                    new_lr = self.model.optimizer.lr * self.reduce_factor
                    self.model.optimizer.lr.assign(new_lr)
                    print("\nEpoch %05d: Reducing learning rate to %s." % (epoch, new_lr))
        else:
            self.wait = 0
            self.stored_epoch = -1

# Callback function for early stopping
ResNet50_BiLSTM_callback = keras.callbacks.EarlyStopping(monitor = 'val_accuracy', patience = 10, verbose = 1)

# Custom-based learning rate scheduler function for reducing learning rate
custom_lr_scheduler = Custom_Learning_Rate_Scheduler(patience = 10, reduce_factor = 0.1, reduce_every = 5)

# Train the ResNet50-BiLSTM model with optimal hyperparameters
history_ResNet50_BiLSTM_model_tuned = ResNet50_BiLSTM_model_tuned.fit(train_generator_resnet50_BiLSTM,
                                                                      epochs = 300,
                                                                      validation_data = val_generator_resnet50_BiLSTM,
                                                                      callbacks = [ResNet50_BiLSTM_callback, custom_lr_scheduler])
```
